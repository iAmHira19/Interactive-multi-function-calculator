```python
import sympy as sp
from fractions import Fraction

def solve_proportion(a, b, c):
    """Solve proportion a/b = c/x for x."""
    x = sp.Symbol('x')
    equation = sp.Eq(a/b, c/x)
    solution = sp.solve(equation, x)
    return solution[0]

def solve_equation(a, b, c):
    """Solve linear equation ax + b = c for x."""
    x = sp.Symbol('x')
    equation = sp.Eq(a*x + b, c)
    solution = sp.solve(equation, x)
    return solution[0]

def factor_square_root(n):
    """Factor the square root of n into its prime factors."""
    root = sp.sqrt(n)
    factors = sp.factorint(n)
    return root, factors

def decimal_to_fraction(decimal):
    """Convert a decimal to a fraction."""
    fraction = Fraction(decimal).limit_denominator()
    return fraction

def decimal_to_percent(decimal):
    """Convert a decimal to a percent."""
    percent = decimal * 100
    return percent

def fraction_to_decimal(fraction):
    """Convert a fraction to a decimal."""
    decimal = float(fraction)
    return decimal

def fraction_to_percent(fraction):
    """Convert a fraction to a percent."""
    decimal = fraction_to_decimal(fraction)
    percent = decimal_to_percent(decimal)
    return percent

def percent_to_decimal(percent):
    """Convert a percent to a decimal."""
    decimal = percent / 100
    return decimal

def percent_to_fraction(percent):
    """Convert a percent to a fraction."""
    decimal = percent_to_decimal(percent)
    fraction = decimal_to_fraction(decimal)
    return fraction

# Main program
def main():
    while True:
        print("\nMulti-Function Calculator")
        print("1. Solve Proportion")
        print("2. Solve Linear Equation")
        print("3. Factor Square Root")
        print("4. Convert Decimal")
        print("5. Convert Fraction")
        print("6. Convert Percent")
        print("7. Exit")

        choice = input("Select an operation (1-7): ")

        if choice == '1':
            print("\nSolve Proportion")
            a = float(input("Enter value for a: "))
            b = float(input("Enter value for b: "))
            c = float(input("Enter value for c: "))
            x_solution = solve_proportion(a, b, c)
            print(f"Solution to proportion {a}/{b} = {c}/x: {x_solution}")

        elif choice == '2':
            print("\nSolve Linear Equation")
            a = float(input("Enter value for a: "))
            b = float(input("Enter value for b: "))
            c = float(input("Enter value for c: "))
            x_solution = solve_equation(a, b, c)
            print(f"Solution to equation {a}x + {b} = {c}: {x_solution}")

        elif choice == '3':
            print("\nFactor Square Root")
            n = float(input("Enter a number to factor its square root: "))
            root, factors = factor_square_root(n)
            print(f"Square root of {n}: {root}")
            print(f"Prime factors of {n}: {factors}")

        elif choice == '4':
            print("\nConvert Decimal")
            decimal = float(input("Enter a decimal value: "))
            fraction = decimal_to_fraction(decimal)
            percent = decimal_to_percent(decimal)
            print(f"Decimal {decimal} as fraction: {fraction}")
            print(f"Decimal {decimal} as percent: {percent}%")

        elif choice == '5':
            print("\nConvert Fraction")
            numerator = int(input("Enter the numerator of the fraction: "))
            denominator = int(input("Enter the denominator of the fraction: "))
            fraction = Fraction(numerator, denominator)
            decimal = fraction_to_decimal(fraction)
            percent = fraction_to_percent(fraction)
            print(f"Fraction {fraction} as decimal: {decimal}")
            print(f"Fraction {fraction} as percent: {percent}%")

        elif choice == '6':
            print("\nConvert Percent")
            percent = float(input("Enter a percent value: "))
            decimal = percent_to_decimal(percent)
            fraction = percent_to_fraction(percent)
            print(f"Percent {percent}% as decimal: {decimal}")
            print(f"Percent {percent}% as fraction: {fraction}")

        elif choice == '7':
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please select a number between 1 and 7.")

# Run the main program
if __name__ == "__main__":
    main()
```
